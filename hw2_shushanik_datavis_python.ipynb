```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_lung = pd.read_csv('/content/lung_cancer_prediction_dataset.csv')
```

```python
df_lung.head()
```

```python
df_lung.info()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(y=df_lung["Annual_Lung_Cancer_Deaths"], color="skyblue")
plt.title("Boxplot of Lung Cancer Deaths Distribution")
plt.ylabel("Annual Lung Cancer Deaths")
plt.show()
```

```python
df_air = pd.read_csv('/content/global_air_pollution_dataset.csv')
```

```python
df_air.head()
```

```python
df_air.info()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_air["PM2.5_AQI_Value"], bins=30, kde=False, color="blue")
plt.title("Histogram of PM2.5 AQI Values")
plt.xlabel("PM2.5 AQI Value")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df_lung["Mortality_Rate"], fill=True, color="red")
plt.title("Density Plot of Lung Cancer Mortality Rate")
plt.xlabel("Mortality Rate")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
fig = plt.gcf()
fig.patch.set_facecolor("#fcfdfe")

ax = plt.gca()
ax.set_facecolor("#dce3f5")

hist = sns.histplot(df_air["PM2.5_AQI_Value"], bins=30, color="blue", edgecolor="black", alpha=0.5, stat="density", label="Histogram")
kde = sns.kdeplot(df_air["PM2.5_AQI_Value"], color="darkred", fill=True, alpha=0.3, linewidth=2, label="Density Plot (KDE)")

plt.title("PM2.5 AQI Distribution with Density Overlay", fontsize=14, fontweight="bold")
plt.xlabel("PM2.5 AQI Value", fontsize=12)
plt.ylabel("Density", fontsize=12)

handles, labels = plt.gca().get_legend_handles_labels()
new_order = [1, 0]
plt.legend([handles[i] for i in new_order], [labels[i] for i in new_order])

plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.figtext(0.6, -0.05, "This plot represents the distribution of PM2.5 AQI values with a density overlay.", wrap=True, ha='right' ,fontsize=10)

plt.show()
```
